```python
import imp
import gensim
from sympy import true
from nltk.corpus import stopwords
from gensim.utils import simple_preprocess
```

## Import OJS text file

```python
ojs = open("ojs_docs.txt").read().split('\n')
```

### Import English stop words

```python
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])
```

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(simple_preprocess(str(sentence), deacc=True))

def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_words] for doc in texts]
```

### convert ojs sentences to words

```python
ojs_words = list(sent_to_words(ojs))
```

### remove stop words in ojs words list

```python
ojs_words_nonstop = remove_stopwords(ojs_words)
```

## Create dictionary and corpus

```python
import gensim.corpora as corpora
# create dictionary
id2word = corpora.Dictionary(ojs_words_nonstop)
# create corpus
texts = ojs_words_nonstop
# term document frequency
corpus = [id2word.doc2bow(text) for text in texts]
```

# LDA model training

```python
from pprint import pprint
# number of topics 
num_topics = 50
lda_model = gensim.models.LdaMulticore(corpus=corpus,id2word=id2word,num_topics=num_topics)
```

```python
doc_lda = lda_model[corpus]
```

```python
def get_topic_matrix(num_topics, num_words, lda_model):
    matrix = np.empty(shape=(num_topics,num_words))
    for idx, topic in lda_model.show_topics(formatted=False, num_topics=num_topics, num_words= num_words):
        # topic contains some tuples of (word, probabilities)
        matrix[idx] = [w[0] for w in topic]
    return matrix
```

```python
topics_25 = lda_model.show_topics(formatted=False, num_topics=50, num_words= 25)
```

```python
def get_topic_list(topics):
    lst = []
    for idx, topic in topics:
        # topic contains some tuples of (word, probabilities)
        lst.append([w[0] for w in topic])
    return lst
```

```python
ojs_topics_list = get_topic_list(topics_25)
```

```python
ojs_topics_matrix = np.asarray(ojs_topics_list)
```

# Compute Topic Diversity

```python
def get_topic_diversity(matrix, topk):
    num_topics = matrix.shape[0]
    list_w = np.zeros((num_topics, topk))
    for k in range(num_topics):
        idx = matrix[k,:].argsort()[-topk:][::-1]
        list_w[k,:] = idx
    n_unique = len(np.unique(list_w))
    TD = n_unique / (topk * num_topics)
    print('Topic diveristy is: {}'.format(TD))
```

```python
get_topic_diversity(ojs_topics_matrix, 25)
```
